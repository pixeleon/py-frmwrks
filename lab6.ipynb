```python
import numpy as np
import pandas as pd
```

```python
books = pd.read_csv("data/books.csv")
books = books.fillna("")
books
```

```python
#task 2
books["TitleNunmers"] = books["Title"].str.findall("\d+").str.join("+")
books[50:76]
```

```python
#task 3
books[books["Publisher"].str.count("[AEIOUaeiou]") < 2] 
```

```python
#task 4
books[books["Author"].str.extract("(\w+)")[0].fillna("").str.match("^[^AEIOU]*[^aeiou]$")]
```

```python
#task 5
books["TitleVowels"] = books["Title"].str.count("[^AEIOUaeiou]")
books.sort_values("TitleVowels").tail(1)
```

```python
#task 6
books[books["Title"].str.count("[A-Z][a-z]*") == 2]
```

```python
#task 7
books[books["Author"].str.extract("(\w+)(, )(\w+)(\w+)")[2].fillna("").str.contains("[^AEIOUaeiou]{3}")]
```

```python
#task 8
books["LastNameLetters"] = books["Author"].str.extract("(\w+)(, )(\w+)(\w+)")[0].str.count("\w")
books[books["LastNameLetters"] == 5]["Author"]
```
